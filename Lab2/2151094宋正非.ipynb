本次作业使用魔塔社区“语音合成-中文-多情感领域-16k-多发音人”中的代码范例。
链接：https://www.modelscope.cn/models/iic/speech_sambert-hifigan_tts_zh-cn_16k

使用代码范例实现中文文字的语音合成时，需要注意使用3.10版本或以下的python。否则会出现报错module 'ttsfrd' has no attribute 'TtsFrontendEngine'，原因是python3.11版本之后ttsfrd团队不维护了。

```python
from modelscope.outputs import OutputKeys
from modelscope.pipelines import pipeline
from modelscope.utils.constant import Tasks

text = '白日不到处，青春恰自来。苔花如米小，也学牡丹开。'
model_id = 'damo/speech_sambert-hifigan_tts_zh-cn_16k'
sambert_hifigan_tts = pipeline(task=Tasks.text_to_speech, model=model_id)
output = sambert_hifigan_tts(input=text, voice='zhitian_emo')
wav = output[OutputKeys.OUTPUT_WAV]
with open('output.wav', 'wb') as f:
    f.write(wav)
```
